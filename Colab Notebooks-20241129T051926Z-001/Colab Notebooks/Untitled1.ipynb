```python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"

# Send GET request to the URL
response = requests.get(url)
html_content = response.text

# Parse the content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Extract the main content of the page (usually inside the <div> with id 'bodyContent')
content = soup.find('div', {'id': 'bodyContent'}).text

# Print a preview of the extracted content
print(content[:1000])  # Display first 1000 characters for preview
```

```python
import re

# Function to clean text
def clean_text(text):
    # Remove references (e.g., [1], [citation needed])
    text = re.sub(r'\[.*?\]', '', text)

    # Remove extra spaces or newline characters
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Clean the text
cleaned_text = clean_text(content)

print(cleaned_text[:1000])  # Display the cleaned text preview
```

```python
# Remove tables
for table in soup.find_all('table'):
    table.decompose()

# Remove all HTML tags
cleaned_text = soup.get_text()

# Optionally remove excessive whitespaces
cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()

print(cleaned_text[:1000])  # Display the cleaned text preview
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Tokenize the cleaned text into words
words = word_tokenize(cleaned_text)

# Remove stopwords (common words like 'the', 'is', etc.)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]

# Print a preview of the filtered words
print(filtered_words[:100])  # Display the first 100 words
```

```python
with open('cleaned_wikipedia_text.txt', 'w', encoding='utf-8') as file:
    file.write(cleaned_text)
```
